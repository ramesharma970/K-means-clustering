```python
import numpy as np
import matplotlib.pyplot as plt

# Function to implement K-Means clustering
def kmeans(X, k, max_iters=100, tol=1e-4):
    # Randomly initialize centroids
    centroids = X[np.random.choice(X.shape[0], k, replace=False)]
    
    for _ in range(max_iters):
        # Step 1: Assign labels based on closest centroid
        distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)
        
        # Step 2: Recompute centroids (mean of points in each cluster)
        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])
        
        # Step 3: Check for convergence (centroid changes)
        if np.all(np.abs(new_centroids - centroids) < tol):
            break
        
        centroids = new_centroids
    
    return centroids, labels

# Generate some random data (for illustration)
np.random.seed(42)
X1 = np.random.randn(100, 2) + [3, 3]
X2 = np.random.randn(100, 2) + [-3, -3]
X3 = np.random.randn(100, 2) + [6, -6]
X = np.vstack([X1, X2, X3])

# Apply K-Means
k = 3
centroids, labels = kmeans(X, k)

# Plot the original data and the clustering result
plt.figure(figsize=(8, 6))




```
